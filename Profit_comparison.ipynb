```python
profit_data = pd.read_csv('Dataset/profit.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np

data_1 = list(profit_data['APPT_XGBoost'])
data_2 = list(profit_data['APPT_ZIP'])


profit_all_data = [data_1, data_2] 
labels=['XGBoost', 'ZIP']

fig, ax = plt.subplots(figsize=(4, 5))

# rectangular box plot
boxplot = ax.boxplot(profit_all_data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels,
                     widths = 0.3)  # will be used to label x-ticks

ax.set_title('Balanced Group Test')
plt.xlabel('Trader')
plt.ylabel('APPT')
plt.show()
```
